# Matplotlib & Seaborn for Data Visualization

```python
import pandas as pd

# Load dataset
df = pd.read_csv("Data Sample Storage/customers-100.csv")
df.head()
```

# Matplotlib 

Matplotlib ek plotting library hai Python mein graphs banane ke liye.

Line chart, bar chart, pie chart—sab kuch visualize kar sakte ho easily.

Ye customization ke liye kaafi flexible hai—colors, labels, axes sab control kar sakte ho.

Scientific aur basic visualizations dono ke liye useful hai.

Thoda verbose lag sakta hai kabhi, par powerful hai analysis ke liye.

```python
 # 1. Line Plot (Basic)
# "Time ke sath kisi bhi cheez ke trend ko dekhne ke liye use hota hai."
import matplotlib.pyplot as plt

x = range(1, 11)
y = [i**2 for i in x]

plt.plot(x, y)
plt.title("Line Plot of x vs x^2")
plt.xlabel("x")
plt.ylabel("x squared")
plt.grid(True)
plt.show()
```

```python
# 🔹 2. Bar Plot (Using Country Count)
# "Alag-alag categories ke comparison ke liye use hota hai (jaise kitne customers kis country mein hain)."


country_count = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(country_count.index, country_count.values, color='skyblue')
plt.title("Number of Customers by Country")
plt.xticks(rotation=90)
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
# 🔹 3. Pie Chart (Top 5 Cities)
# "Total ka kitna percentage har category ka hai yeh dikhane ke liye use hota hai."
top_cities = df['City'].value_counts().nlargest(5)

plt.pie(top_cities.values, labels=top_cities.index, autopct='%1.1f%%')
plt.title("Top 5 Cities by Customer Count")
plt.show()
```

```python
# 4. Horizontal Bar (Advanced Custom)
# "Categorical data mein har value kitni baar aayi hai yeh dekhne ke liye use hota hai."
plt.barh(top_cities.index, top_cities.values, color='green')
plt.title("Top 5 Cities (Horizontal Bar)")
plt.xlabel("Customers")
plt.show()
```

# Seaborn

Seaborn Matplotlib ke upar built hai—aur easy aur beautiful plots deta hai.

Statistical plots jaise histograms, boxplots, violin plots banane mein best hai.

DataFrame ke saath kaam karna super smooth hai isme.

Themes aur color palettes se graphs visually attractive ho jaate hain.

Exploratory data analysis mein Seaborn kaafi helpful hota hai.

```python
import seaborn as sns
```

```python
# Countplot (Country)
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Country', order=df['Country'].value_counts().index)
plt.title("Customer Count by Country")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 2. Boxplot (Length of First Names by Country)
# "Kisi numerical data ka spread, median aur outliers ko samajhne ke liye use hota hai."
df['Name Length'] = df['First Name'].apply(len)

plt.figure(figsize=(18,6)) # size of Figure
sns.boxplot(data=df, x='Country', y='Name Length')
plt.title("Name Length Distribution by Country")
plt.xticks(rotation=90)
plt.show()
```

```python
#  3. Histogram / Distplot
# "Numerical data ka distribution dekhne ke liye use hota hai – kitne values kis range mein hain."
sns.histplot(df['Name Length'], bins=10, kde=True)
plt.title("Distribution of First Name Lengths")
plt.xlabel("Name Length")
plt.show()
```

2. Line Plot & Horizontal Bar Plot

Question:

Using the customers-100.csv dataset:

(a) Plot a line chart showing the cumulative number of customers added as you go down the dataset (assume each row is a new customer).

(b) Create a horizontal bar plot showing the number of customers in the top 5 cities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Data Sample Storage/customers-100.csv')
```

```python
# we track cumulative customers by counting rows progressively.
df['Customer Count'] = range(1, len(df) + 1)
```

```python
# Progressive customer count
# for showing customer growth over time.
```

```python
plt.plot(df['Customer Count'], label='Cumulative Customers', color='blue')
plt.xlabel('Row Number')
plt.ylabel('Cumulative Customers')
plt.title('Cumulative Number of Customers Over Time')
plt.grid(True)
plt.legend() # Cumulative Customer
plt.tight_layout()
plt.show()
```

```python
# Count top 5 cities
top_cities = df['City'].value_counts().head(5)

# Horizontal bar plot
plt.figure(figsize=(5, 3))
plt.barh(top_cities.index, top_cities.values, color='green')
plt.xlabel('Number of Customers')
plt.title('Top 5 Cities by Number of Customers')
plt.tight_layout()
plt.show()
```
